# Bad Output

This network demonstrates that bad output is checked with the output bank's activation function.

```python
import conx as cx
```

We make a simple network, using the "sigmoid" activation function on the hidden and output layers.

```python
net = cx.Network("Bad Output", 2, 3, 1, activation="sigmoid")
```

```python
net.compile(error="mse", optimizer="sgd")
```

If we see a picture of the network at this point, and mouse-over the input bank, we see that the ranges of values are shown to be -Infinity, to +Infinity.

```python
net.picture()
```

If we use a target in range, no problem:

```python
net.dataset.append([0, 1], [1])
```

However, if we give a target out of range, we get a warning:

```python
net.dataset.append([1, 0], [-1])
```

Also note that a new picture of the network now reflects the dataset given. That is, a mouse-over of the input bank shows a range of 0.0 to 1.0.

```python
net.picture()
```
