# Utilities

This notebook demonstrates the Conx utility functions.

First, we import the conx library:

```python
import conx as cx
net = cx.Network("XOR Network", 2, 4, 1, activation="sigmoid")
```

## Images

**array2image(array, scale=1.0, shape=None)**

```python
r, g, b = 0, 0, 1
array = [[[r,g,b] for col in range(400)] for row in range(200)]
image = cx.array2image(array)
image
```

**image2array(image)**

```python
cx.shape(cx.image2array(image))
```

```python
cx.image2array(image)[0][0]
```

**svg2image(svg, background=(255, 255, 255, 255))**

```python
svg = net._repr_svg_()
image = cx.svg2image(svg)
image
```

**crop_image(image, x1, y1, x2, y2)**

```python
cx.crop_image(image, 10, 10, 200, 200)
```
